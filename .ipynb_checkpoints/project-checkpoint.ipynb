## About the method

In class, we've discussed that "Splines are a way to guarantee an arbitrary amount of smoothness". 

"Patsy offers a set of specific stateful transforms that you can use in formulas to generate splines bases and express non-linear fits." (https://patsy.readthedocs.io/en/latest/spline-regression.html)

Patsy includes b-splines, natural splines, and cyclic cubic regression splines. From class, we've discussed that "B-splines are a more computationally convenient representation of a piecewise polynomial curve".

Here is documentation about one of the spline regression functions from https://patsy.readthedocs.io/en/latest/API-reference.html#spline-regression
<img src="numcomp_bspline.png"
     style="width:500px;height:600px;" />

The following examples come from: https://patsy.readthedocs.io/en/latest/spline-regression.html

```python
# https://patsy.readthedocs.io/en/latest/spline-regression.html

import matplotlib.pyplot as plt

plt.title("B-spline basis example (degree=3)");

x = np.linspace(0., 1., 100)

y = dmatrix("bs(x, df=6, degree=3, include_intercept=True) - 1", {"x": x})

# Define some coefficients
b = np.array([1.3, 0.6, 0.9, 0.4, 1.6, 0.7])

# Plot B-spline basis functions (colored curves) each multiplied by its coeff
plt.plot(x, y*b);

# Plot the spline itself (sum of the basis functions, thick black curve)
plt.plot(x, np.dot(y, b), color='k', linewidth=3);
```

```python
#https://patsy.readthedocs.io/en/latest/spline-regression.html

plt.title("Cyclic cubic regression spline basis example");

y = dmatrix("cc(x, df=6) - 1", {"x": x})

# Plot cyclic cubic regression spline basis functions (colored curves) each multiplied by its coeff
plt.plot(x, y*b);

# Plot the spline itself (sum of the basis functions, thick black curve)
plt.plot(x, np.dot(y, b), color='k', linewidth=3);
```

## Open questions

Patsy includes tensor product smooths, and I don't believe we've covered them in class. 

"Tensor product smooths. Smooths of several covariates can be generated through a tensor product of the bases of marginal univariate smooths." (https://patsy.readthedocs.io/en/latest/spline-regression.html)

See the example below using tensor products to represent a smooth of two variables from https://patsy.readthedocs.io/en/latest/spline-regression.html

I'm interested in learning more about how tensor products can be applied in real contexts. Searching online, it seems that tensor product smooths can be utilized in generalized additive models (GAMs). Why are tensor product smooths the best method for estimating smooth relationships? Why do tensor products benefit GAMs? 

```python
from matplotlib import cm
from mpl_toolkits.mplot3d.axes3d import Axes3D
x1 = np.linspace(0., 1., 100)
x2 = np.linspace(0., 1., 100)
x1, x2 = np.meshgrid(x1, x2)
df = 3
y = dmatrix("te(cr(x1, df), cc(x2, df)) - 1",
           {"x1": x1.ravel(), "x2": x2.ravel(), "df": df})

print(y.shape)
fig = plt.figure()
fig.suptitle("Tensor product basis example (2 covariates)");
for i in range(df * df):
    ax = fig.add_subplot(df, df, i + 1, projection='3d')
    yi = y[:, i].reshape(x1.shape)
    ax.plot_surface(x1, x2, yi, rstride=4, cstride=4, alpha=0.15)
    ax.contour(x1, x2, yi, zdir='z', cmap=cm.coolwarm, offset=-0.5)
    ax.contour(x1, x2, yi, zdir='y', cmap=cm.coolwarm, offset=1.2)
    ax.contour(x1, x2, yi, zdir='x', cmap=cm.coolwarm, offset=-0.2)
    ax.set_xlim3d(-0.2, 1.0)
    ax.set_ylim3d(0, 1.2)
    ax.set_zlim3d(-0.5, 1)
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])
    ax.set_zticks([-0.5, 0, 1])

fig.tight_layout()
```
